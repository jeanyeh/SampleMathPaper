```python
from itertools import permutations
########## input k ###############
k = 4
########################

def perfect_matching(k: int):
  for raw in permutations(range(3, k+1), k-2):
    # quick way to generate possible case
    raw = [1, raw[0], 2] + list(raw[1:])

    # find the perfect matching
    for idx in range(0, k-1, 2):
      if raw[idx] > raw[idx+1]:
        # ith > jth
        break
      if idx+2 < k and raw[idx] > raw[idx+2]:
        # ith > i+2 th
        break

      # wonder matching
      x, y = raw.index(idx+1), raw.index(idx+2)
      if x % 2 != y % 2:
        break
    else:
      yield [(raw[n], raw[n+1]) for n in range(0, k, 2)]

idx = 0
parity = []
for mached in perfect_matching(4*k):
  idx += 1
  ee = 0
  eo = 0
  oe = 0
  oo = 0
  for i in range(2*k):
    if ( mached[i][0] %2 ) * ( mached[i][1] %2 ) == 1 :
      oo += 1
    elif ( ( mached[i][0]+1 ) %2 ) * ( mached[i][1] %2 ) == 1 :
      eo += 1
    elif ( mached[i][0] %2 ) * ( ( mached[i][1]+1 ) %2 ) == 1 :
      oe += 1
    else:
      ee += 1
  parity.append([ee, eo, oe, oo])
  print(f"{idx}.   **{mached}**, ee= **{ee}**, eo= **{eo}**, oe= **{oe}**, oo= **{oo}**\n")

for a in range(k+1):
  check = 0
  for i in range(len(parity)):
    if parity[i] == [a, k-a, k-a, a]:
      check += 1
  print(f"For type ee=**{a}**, eo=**{k-a}**, oe=**{k-a}**, oo=**{a}**, we have **{check}** many\n")
```
